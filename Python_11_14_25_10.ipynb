```python
#budeme zadavat text -> binarni cisla

list_bin = [] # list_pro jednotlive znaky
slova = "Python je super."  #zadana veta
list_slov = slova.split(" ") #rozdel na jednotliva slova

for slovo in list_slov: #pro vsechna slova
  for pismeno in slovo: #pro vsechna pismena
    cele_cislo = ord(pismeno) #cislo
    binarka = bin(cele_cislo) #binarka
    list_bin.append(binarka) #naplneni listu

  list_bin.append("0b100000") #pridej mezeru
print(list_bin)
```
